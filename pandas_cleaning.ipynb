```python
import pandas as pd

# Merge HTS Migori and HTS Kakamega data
hts_migori = pd.read_csv('./raw_data/hts_migori.csv')
hts_kakamega = pd.read_csv('./raw_data/hts_kakamega.csv')
hts_merged = pd.concat([hts_migori, hts_kakamega], ignore_index=True)

# Get the column names
# headers = hts_merged.columns.tolist()
# print(headers)

hts_merged.info()
```

```python
hts_merged['datim_value'].fillna(int("0"), inplace=True)
hts_merged.info()
```

```python
# Extract age and gender from Indicator Name in HTS data
hts_merged[['gender']] = hts_merged['Indicator Name'].str.extract(r'(\(F\)|\(M\))')
hts_merged.head(5)
```

```python
hts_merged.isnull().sum()
```

```python
hts_merged = hts_merged.dropna(how='any')
```

```python
hts_merged.isnull().sum()
```

```python
hts_merged.info()
```

```python
# Extract age from Indicator Name in HTS data
hts_merged[['age']] = hts_merged['Indicator Name'].str.extract(r'(?P<age><1|1-9|10-14|15-19|20-24|25\+)')
hts_merged.head(5)
```

```python
# Create short indicator name in HTS data
hts_merged['short_indicator_name'] = hts_merged['Indicator Name'].apply(lambda x: 'tested' if 'Tested_' in x else None)
hts_merged.head(5)
```

```python
# Drop Indicator Name from HTS data
hts_merged = hts_merged.drop(columns=['Indicator Name'])
hts_merged.head(5)
```

```python
# Drop Indicator Name from HTS data
hts_merged = hts_merged.drop(columns=['short_indicator_name'])
hts_merged.head(5)
```
